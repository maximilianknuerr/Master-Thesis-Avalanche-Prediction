```python
mooser_wetter_daten = pd.read_csv('data/Mooser_Wetter_Daten.csv')
mooser_wetter_daten
```

```python
mooser_wetter_daten['Datum'] = mooser_wetter_daten['Datum'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') if type(x)==str else pd.NaT)
mooser_wetter_daten = mooser_wetter_daten.loc[(mooser_wetter_daten['Datum'] > start) & (mooser_wetter_daten['Datum'] < end)]
mooser_wetter_daten.reset_index(inplace=True, drop=True)
mooser_wetter_daten
```

```python
import statistics
from math import isnan
for i in mooser_wetter_daten.index:
    j = i - 1
    Neuschnee_last2 = 0.0
    Neuschnee_last4 = 0.0


    windstaerke_last4 = []
    Lufttemperatur_last4 = []

    while mooser_wetter_daten.iloc[j]['Intervall'] == mooser_wetter_daten.iloc[i]['Intervall'] and j  >= 0  and j >= i - 10:
        if j >= i - 2:
            Neuschnee_last2 += mooser_wetter_daten.iloc[j]['Neuschnee_Norm']
        if j >= i - 4:
            Neuschnee_last4 += mooser_wetter_daten.iloc[j]['Neuschnee_Norm']
            windstaerke_last4.append(mooser_wetter_daten.iloc[j]['Windstärke_Norm'])
            Lufttemperatur_last4.append([mooser_wetter_daten.iloc[j]['Lufttemperatur_7h_Norm'], mooser_wetter_daten.iloc[j]['Lufttemperatur_14h_Norm'], mooser_wetter_daten.iloc[j]['Lufttemperatur_19h_Norm']])
        j -= 1

    mooser_wetter_daten.loc[i, 'Neuschnee_last4'] = Neuschnee_last4
    mooser_wetter_daten.loc[i, 'Neuschnee_last2'] = Neuschnee_last2
    if len(windstaerke_last4) > 0:
        for j in range(4):
            if len(windstaerke_last4) > j:
                mooser_wetter_daten.loc[i, 'windstaerke_last' + str(j + 1)] = windstaerke_last4[j]
            else:
                mooser_wetter_daten.loc[i, 'windstaerke_last' + str(j + 1)] = 0.0
    else:
        for j in range(4):
            mooser_wetter_daten.loc[i, 'windstaerke_last' + str(j + 1)] = 0.0
    if len(Lufttemperatur_last4) > 0:
        for j in range(4):
            if len(Lufttemperatur_last4) > j:
                mooser_wetter_daten.loc[i, 'Lufttemperatur_last' + str(j + 1)] = statistics.mean(Lufttemperatur_last4[j])
            else:
                mooser_wetter_daten.loc[i, 'Lufttemperatur_last' + str(j + 1)] = 0.0
    else:
        for j in range(4):
            mooser_wetter_daten.loc[i, 'Lufttemperatur_last' + str(j + 1)] = 0.0




mooser_wetter_daten
```
