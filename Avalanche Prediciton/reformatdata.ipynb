```python
import numpy as np
import pandas as pd
```

```python
lawinen = pd.read_csv('data/Allg_Lawinen_Abgänge.csv')
```

```python
import datetime as dt

lawinen['DATUM'] = lawinen['DATUM'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') if type(x)==str else pd.NaT)
lawinen['ZEIT'] = lawinen['ZEIT'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') if type(x)==str else pd.NaT)
lawinen
```

```python
lawinen.sort_values(by=['DATUM','ZEIT'],inplace=True)
lawinen
```

```python
start = dt.datetime.strptime('1992-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
end = dt.datetime.strptime('2022-12-31 00:00:00', '%Y-%m-%d %H:%M:%S')
lawinen = lawinen.loc[(lawinen['DATUM'] > start) & (lawinen['DATUM'] < end)]
lawinen
```

```python
kaplawstr = pd.read_csv('data/kaplawstr.csv' ,sep=';')
#lawinen = pd.merge(lawinen, lawinen['Lawinen_Name'].apply(lambda x: kaplawstr.loc[kaplawstr['Lawinenstrich']==x]), left_index=True, right_index=True)
#lawinen
lawinen = pd.merge(lawinen, kaplawstr, left_on='GRNR', right_on='ID')
lawinen
```

```python
TOPP = pd.read_csv('data/TOPP.csv')
TOPP
```

```python
TOPP = kaplawstr['Code_neu'].apply(lambda x: TOPP.loc[TOPP['Lawinencode'] == x].mean())

#TOPP = TOPP.groupby('Lawinencode').first().reset_index()
```

```python
lawinen = pd.merge(lawinen, TOPP, left_on='Code_neu', right_on='Lawinencode')
lawinen
```

```python
lawinen.to_csv('newData/lawinen_abgänge.csv', index=True)
```

```python
mooser_wetter_daten = pd.read_csv('data/Mooser_Wetter_Daten.csv')
mooser_wetter_daten
```

```python
mooser_wetter_daten['Datum'] = mooser_wetter_daten['Datum'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S') if type(x)==str else pd.NaT)
mooser_wetter_daten = mooser_wetter_daten.loc[(mooser_wetter_daten['Datum'] > start) & (mooser_wetter_daten['Datum'] < end)]
```

```python
lawinen['DATUM'] = pd.to_datetime(lawinen['DATUM'])


gesamt_df = pd.merge(mooser_wetter_daten, lawinen, right_on='DATUM', left_on='Datum', how='outer')
gesamt_df
```

```python
#gesamt_df['meanExpo'].replace(np.nan, TOPP.sample(1)['meanExpo'].values[0], inplace=True)

for i in gesamt_df.index:
    if(pd.isnull(gesamt_df['meanExpo'][i])):
        sample = TOPP.sample(1)
        gesamt_df['meanExpo'][i] = sample['meanExpo'].values[0]
        gesamt_df['meanSlope'][i] = sample['meanSlope'].values[0]
        gesamt_df['stdDevSlope'][i] = sample['stdDevSlope'].values[0]
        gesamt_df['MinSlope'][i] = sample['MinSlope'].values[0]
        gesamt_df['MaxSlope'][i] = sample['MaxSlope'].values[0]
        gesamt_df['Altitude'][i] = sample['Altitude'].values[0]
gesamt_df
```
